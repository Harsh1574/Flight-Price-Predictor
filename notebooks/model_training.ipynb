```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
df = pd.read_csv('E:/Machine Learning/Flight Price Predictor/datasets/flight_price_cleaned.csv')
df.head()
```

```python
df.info()
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.drop('Price', axis=1), df['Price'], test_size=0.2, random_state=42)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
linear = LinearRegression()
scaler = StandardScaler()
# x_train = scaler.fit_transform(x_train)
# x_test =scaler.transform(x_test)
```

```python
linear.fit(x_train, y_train)
linear_prediction = linear.predict(x_test)
```

```python
mae = mean_absolute_error(y_test, linear_prediction)
mse = mean_squared_error(y_test, linear_prediction)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, linear_prediction)
print(f'Mean Absolute Error: {mae}')
print(f'Root Mean Squared Error: {rmse}')
print(f'R^2 Score: {r2}')
```

```python
from sklearn.svm import SVR
svr=SVR()
```

```python
svr.fit(x_train, y_train)
```

```python
svr_prediction = svr.predict(x_test)
mae = mean_absolute_error(y_test, svr_prediction)
mse = mean_squared_error(y_test, svr_prediction)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, svr_prediction)
print(f'Mean Absolute Error: {mae}')
print(f'Root Mean Squared Error: {rmse}')
print(f'R^2 Score: {r2}')
```

```python
import pickle
pickle.dump(linear, open('flight_price_linear_model.pkl','wb'))
```

```python
kernel= ["rbf"]
gamma =[0.01, 0.1, 1.0]
C = [0.1, 1, 10]
params=dict(kernel=kernel, gamma=gamma, C=C)
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(estimator=SVR(), param_grid=params)
```

```python
grid.fit(x_train, y_train)
print("Best Parameters:", grid.best_params_)
print("Best Score:", grid.best_score_)
```

```python
grid_prediction = grid.predict(x_test)
mae = mean_absolute_error(y_test, grid_prediction)
mse = mean_squared_error(y_test, grid_prediction)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, grid_prediction)
print(f'Mean Absolute Error: {mae}')
print(f'Root Mean Squared Error: {rmse}')
print(f'R^2 Score: {r2}')
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

ridge = Ridge()
params = {'alpha': [0.1, 1.0, 10.0, 100.0]}
grid = GridSearchCV(ridge, params, cv=5, scoring='r2')
grid.fit(x_train, y_train)
print("Best alpha:", grid.best_params_)
print("Best score:", grid.best_score_)
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV

lasso = Lasso()
params = {'alpha': [0.1, 1.0, 10.0, 100.0]}
grid_lasso = GridSearchCV(lasso, params, cv=5, scoring='r2')
grid_lasso.fit(x_train, y_train)
print("Best alpha:", grid_lasso.best_params_)
print("Best score:", grid_lasso.best_score_)
```

```python
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import GridSearchCV

lasso = ElasticNet()
params = {'alpha': [0.1, 1.0, 10.0, 100.0]}
grid = GridSearchCV(lasso, params, cv=5, scoring='r2')
grid.fit(x_train, y_train)
print("Best alpha:", grid.best_params_)
print("Best score:", grid.best_score_)
```

```python
pickle.dump(grid_lasso, open('flight_price_lasso_model.pkl','wb'))
```
